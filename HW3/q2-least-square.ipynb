```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
X_and_Y = np.load('./q2-least-square.npy')
X = X_and_Y[:, 0]  # Shape: (300,)
Y = X_and_Y[:, 1]  # Shape: (300,)
```

### Original Data

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(X, Y)
plt.show()
```

### Linear Regression

```python
# Assume Y = w0 + w1 * X = (w0, w1).(1, X) = W.X1
X1 = np.matrix(np.hstack((np.ones((len(X),1)),
                          X.reshape(-1,1))))  # X1 contains 1 and X.
W = X1.T.dot(X1).I.dot(X1.T).dot(Y)
w0, w1 = np.array(W).reshape(-1)
print('Y = {:.2f} + {:.2f}*X'.format(w0, w1))
```

```python
X_line = np.linspace(0,10,300)
Y_line = w0 + w1 * X_line
plt.scatter(X, Y)
plt.plot(X_line, Y_line, color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Polynomial Regression

```python
# Assume Y = w0 + w1 * X + w2 * X^2 = (w0, w1, w2).(1, X, X^2) = W.X2
X2 = np.matrix(np.hstack((np.ones((len(X),1)),
                          X.reshape(-1,1), 
                          (X**2).reshape(-1,1))))  # X2 contains 1, X and X^2.
W = X2.T.dot(X2).I.dot(X2.T).dot(Y)
w0, w1, w2 = np.array(W).reshape(-1)
print('Y = {:.2f} + {:.2f}*X + {:.2f}*X^2'.format(w0, w1, w2))
```

```python
X_line = np.linspace(0,10,300)
Y_line = w0 + w1 * X_line + w2 * (X_line**2)
plt.scatter(X, Y)
plt.plot(X_line, Y_line, color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
